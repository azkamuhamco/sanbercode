<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
<ul>
    <li>Perbedaan Series dengan list dan dictionary</li>
    Series dapat mengatur index, sedangkan list dan dictionary tidak dapat mengatur index. Index pada list dan dictionary berdasarkan urutan data yang dimasukkan
    <br><br><li>Perbedaan dataframe dan series</li>
    Dataframe = Objek dua dimensi => Data bisa terdiri dari banyak kolom<br>
    Series = Objek satu dimensi => Data hanya bisa satu kolom
</ul>

```python
# code here
x = pd.Series([16,23,34,46,58], index=['o', 'p', 'q', 'r', 's'])
x
```

Di soal no 2 kamu telah membaca data tentang diamonds. pada soal ini cobalah suatu metode dari pandas untuk mengetahui beberapa data pertama dan beberapa data terakhir. kemudian sebutkan nilai dari:

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
print("rata-rata dari column price =", df1['price'].mean())
print("standar deviasi dari column depth =", df1['depth'].std())
print("nilai max dari column carat =", df1['carat'].max())
```
